# Data Visualization:

```python
Merged_dataset = pd.merge(dataset1,dataset2,how="outer", on=["Item Number"])
Merged_dataset
```

1. Create a table showing inventory levels at different dates in the future[Weekly Basis] for each item using pandas.

```python
#Creating empty Dataframe to store the inventory levels for weekly basis:
weekly_basis = pd.date_range(start="6/17/2023", end="8/26/2023", freq="W")# W for weekly 
future_weekly = pd.DataFrame(index=weekly_basis)
```

```python
#Iterating through items and calculating inventory levels for each future date:
for item in Merged_dataset["Item Number"].unique():
    item_data = Merged_dataset[Merged_dataset["Item Number"] == item]
    inventory_levels = []

    for date in weekly_basis:
        total_inventory = item_data.loc[item_data["Date"] <= date, "Inventory"].sum()
        new_inventory = item_data.loc[
            (item_data["Inventory Receive date"] <= date) & (item_data["Inventory Receive date"].notnull()),
            "Total Item Qty"].sum()
        inventory_levels.append(total_inventory + new_inventory)

    future_weekly[item] = inventory_levels
```

```python
print(future_weekly)
```

2. Create a bar chart showing total monthly inventory levels:

```python
#Calculating total monthly inventory levels:
monthly_total = future_weekly.resample('M').sum()
```

```python
#Bar chart for total monthly inventory levels
plt.figure(figsize=(10, 6))
monthly_total.plot(kind="bar", stacked=True)
plt.title("Total Monthly Inventory Levels")
plt.xlabel("Month")
plt.ylabel("Total Inventory")
plt.legend(title="Item Number",bbox_to_anchor=(0,0))
plt.show()
```
